## Problem 2: Iris (3 pts)
Problem 2 is based on the `iris` data frame with 150 rows and 5 columns. The first 6 rows have been displayed in the following cell. You can run `help(iris)` to get the description of the data set.

````r
head(iris)
````

**(a)** Following the example in the lab, reproduce the following density plot for petal length with each iris species. What insights can you derive by looking at this density plot?: _1 points_

p2a.png

````r
ggplot(iris, aes(x = Petal.Length)) +
  geom_density(aes(fill = Species), alpha=0.5) +
  theme_bw() + labs(
    title = "Density plot of petal length across species",
    x = "Petal length (cm)", y = "Species",
    color = "Species")
    By looking at this density plot we can determine that the petal length for all setosa species are shorter than the other two species and while there is slight overlap between the versicolor and virginica, the virginica tends to have longer petals.
````

**(b)** How are petal length and petal width related? Following the example we saw in lecture, reproduce the following plot: _1 points_


p2b.png

````r
ggplot(
  data = iris,
  mapping = aes(x = Petal.Length, y = Petal.Width)
) +
  geom_point(mapping = aes(color = Species, shape = Species)) +
  geom_smooth() +
  labs(
    title = "Petal length and width for iris flowers",
    x = "Petal length (cm)", y = "Petal width (cm)")+
  theme_bw()
As the Petal length increases, so does the Petal length and this relationship is roughly linear.
````

**(c)**
The preceding plot indicates that the petal width is roughly linearly related to the petal length. Check this by computing the correlation of petal width  with petal length, and round the correlation to two decimal places. (*Hint*: R contains a built-in function; for computing the correlation of two vectors try `help(cor)` to learn more; and similarly another built-in function for rounding values to a specific number of decimal places.) _1 point_

````r
data <- cor(iris$Petal.Length, iris$Petal.Width)
round(data, digits = 2)
````
